## Tablas de frecuencias

```python
import random
```

```python
# basado en el LIBRO estad√≠stica para administraci√≥n pagina 12
# creamos una lista vacia que representa la edad de 100 personas

edades= []

# Hacemos un ciclo para generar 100 n√∫meros aleatorios que representan la edad de las personas
for i in range(100):
    # generamos un n√∫mero aleatorio con la funcion randint y lo guardamos en la variable
    # numero_aleatorio
    numero_aleatorio = random.randint( 18, 70)
    # Agregamos el n√∫mero aleatorio a la lista de valores
    edades.append(numero_aleatorio)

print(edades)
```

```python
import pandas as pd
```

```python
# Crear un DataFrame de pandas con los n√∫meros aleatorios
df = pd.DataFrame(edades, columns=['Edad'])
df.head()
```

```python
# Calcular la tabla de distribuci√≥n de frecuencias
tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')

# Imprimir la tabla de distribuciones de frecuencias
print("Tabla de Distribuciones de Frecuencias:")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual'] = tabla_frecuencias['Frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

## Regla de sturges
1. Es una regla que sirve para calcular el n√∫mero de clases o intervalos id√≥neos en los que que se debe dividir un conjunto de datos.
2.Establece que **el n√∫mero de clases es igual a uno m√°s el logaritmo en base dos del n√∫mero total de datos.**

** * C= 1 + log¬≤(N)* **

Donde *N* es el tama√±o de la muestra.

```python
import math
```

```python
# Volvemos a usar la lista de edades
print (edades)
```

```python
# Crear un DataFrame con las edades
df = pd.DataFrame({'Edad': edades})
df
```

```python
# Calcular el n√∫mero de intervalos utilizando la regla de Sturges
n = len(edades)
k = int( 1 + math.log2(100))
print("Valor de n = ", n , " \nN√∫mero de intervalos =", k)
```

```python
# Crear la tabla de frecuencias

tabla_frecuencia = pd.cut( df['Edad'], bins=k, include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns = ['Clase', 'Frec']
tabla_frecuencia
```

```python
# Ordenar la tabla de frecuencia por el intervalo
tabla_frecuencia = tabla_frecuencia.sort_values(by='Clase')
tabla_frecuencia
```

```python
# Generar columnas siguientes
tabla_frecuencia['Frec Rel'] = tabla_frecuencia['Frec'] / n
tabla_frecuencia['Frec Porc'] = tabla_frecuencia['Frec'] / n * 100
tabla_frecuencia['Frec Ac'] = tabla_frecuencia['Frec'].cumsum()
tabla_frecuencia['Frec Rel Ac'] = tabla_frecuencia['Frec Rel'].cumsum()
tabla_frecuencia['Frec Porc Ac'] = tabla_frecuencia['Frec Porc'].cumsum()

# Mostrar la tabla de frecuencia
print(tabla_frecuencia)
```

```python

import matplotlib.pyplot as plt # Importing the matplotlib.pyplot module and assigning it to the alias 'plt'

plt.hist(edades, alpha = 0.5)
plt.title("Histograma de edades basado en la funcion de matplotlib con valores est√°ndar")
```

```python
# Crear un histograma basado en la tabla de frecuencia
plt.hist(df['Edad'], bins=k, alpha = 0.5)
plt.title('Histograma de edades basado en la regla de Sturges')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
```

#   **Pr√°ctica #5** *Diagramas de Cajas y bigotes.*

```python



edades2= []

# Hacemos un ciclo para generar 100 n√∫meros aleatorios que representan la edad de las personas
for i in range(100):
    # generamos un n√∫mero aleatorio con la funcion randint y lo guardamos en la variable
    # numero_aleatorio
    numero_aleatorio = random.randint( 20, 87)
    # Agregamos el n√∫mero aleatorio a la lista de valores
    edades2.append(numero_aleatorio)

print(edades2)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

bplot1 = ax1.boxplot(edades,
                     vert=True, # vertical box alignment
                     patch_artist=True) #fill with color
ax1.set_title('Edades Encuenta 1')

bplot2 = ax2.boxplot(edades2,
                     vert=True, # vertical box alignment
                     patch_artist=True) #fill with color
ax2.set_title('Edades Encuenta 2')
plt.show()
```
