```python
# Se importan todas las librerias que ocupamos
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
import matplotlib.pyplot as plt
from statistics import *
import math
```

```python
# Cargar el archivo CSV en una variable llamada Lobos
Lobos = pd.read_csv("Medicina.csv", encoding = "ISO-8859-1")
```

```python
Lobos.head() #Muestra los primeros 5 lineas del documento
```

# **Valores Maximos y minimos**

```python
Lobos.Horas.max()
```

```python
Lobos.Semestre.max()
```

```python
Lobos.Horas.min()
```

```python
Lobos.Semestre.min()
```

# **Media**

```python
horas = Lobos['Horas'].dropna()

mediaH = sum(horas) / len(Lobos.Horas)
print(" Media de las horas: ", mediaH)

Semestre = Lobos['Semestre'].dropna()

mediaS = sum(Semestre) / len(Lobos.Semestre)
print(" Media de los semestres: ", mediaS)
```

# **Mediana**

```python
medianaH = median(Lobos.Horas)
print("Mediana de las horas: ", medianaH)

medianaS = median(Lobos.Semestre)
print("Mediana de los semestres: ", medianaS)
```

# **Moda**

```python
# Si solo hay una moda se usa este
modaH = mode(Lobos.Horas)
print(modaH)

# Si hay mas de una moda se usa este
modaH = multimode(Lobos.Horas)
print("Moda de las horas: ", modaH)

modaS = multimode(Lobos.Semestre)
print("Moda de los semestres: ", modaS)
```

# **Rango**

```python
# El data frame se usa para inicilizar todos los valores (en este caso,
# las horas) para despues usarlos.
dfH = pd.DataFrame({'Horas': Lobos.Horas})
dfH
```

```python
# El data frame se usa para inicilizar todos los valores (en este caso,
# los semestres) para despues usarlos.
dfS = pd.DataFrame({'Semestre': Lobos.Semestre})
dfS
```

```python
rangoH = dfH["Horas"].max() - dfH["Horas"].min()
print("Rango de las horas:", rangoH)
```

```python
rangoS = dfS["Semestre"].max() - dfS["Semestre"].min()
print("Rango de los semestres:", rangoS)
```

# **Varancia**

```python
varianzaH = dfH["Horas"].var()
print("Varianza de las horas:", varianzaH)
```

```python
varianzaS = dfS["Semestre"].var()
print("Varianza de las semestre:", varianzaS)
```

# **Desviación estandar**

```python
desviacion_estandar_horas = Lobos.Horas.std()
print("Desviación estandar de las horas: ", desviacion_estandar_horas)

desviacion_estandar_sem = Lobos.Semestre.std()
print("Desviación estandar de los semestres: ", desviacion_estandar_sem)
```

# **Histograma**

```python
plt.hist(dfH["Horas"], bins=5, edgecolor="black")
plt.xlabel("Horas")
plt.ylabel("Frecuencia")
plt.title("Histograma de Horas")
plt.show()
```

```python
plt.hist(dfS["Semestre"], bins=5, edgecolor="black")
plt.xlabel("Semestres")
plt.ylabel("Frecuencia")
plt.title("Histograma de los semestres")
plt.show()
```

# **Caja de bigotes (extra)**

```python
sns.boxplot(x='Horas', y='Semestre', data = Lobos)
plt.title('Distribución de las horas por semestres', fontsize=14)
plt.xlabel('Horas estudiadas', fontsize=12)
plt.ylabel('Semestre', fontsize=12)
plt.show()
```

Se llama caja de bigotes, se usa para comparaciones, haz de cuenta que las cajas azules son donde se concentran los datos siendo las lineas negras la media de cada una. Las bolitas son valores atipicos, es decir valores que no son del promedio (por ejemplo, el wey de 10mo semestre que estudio 3 horas)
